Definimos la función para generar el cero y la pendiente

```python
def my_ceropen(x,ymin = -5, ymax = 5): #Definimos la función ceropen para tomar la pendiente de cada punto discretizado en la señal
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin) #pendiente de la recta a partir de 2 puntos
  c = ymin - m*xmin #punto de intersección de la recta
  yv = m*x + c # se le asigna a yv la ecuacion de la recta análoga a y = mx + b
  return yv #retornamos la recta
```

Ahora, definimos la función para cuantizar punto a punto

```python
def my_cuantizador(yn, vq) : # yn punto a cuantizar, vq vector de estados

  Ne = vq.shape[0] #tamaño vector de estados
  dn = cdist(yn.reshape(-1,1),vq.reshape(-1,1)) #distancia yn a vector de estados, reshape(-1.1), asegura vectores columna para poder utilizar cdist
  ind = np.argmin(dn) #posición distancia minima
  return vq[ind] #retorna un nuevo vector con ind
```

Realizamos la discretización de la señal de acuerdo a la fecuencia de muestro, en este caso es de $Fs=5000Hz$

```python
import numpy as np  #importamos la libreria numpy con el alis de np, para que reciba los arreglos
import matplotlib.pyplot as plt  #importamos la libreria de matplotlib.pyplot con el alias de plt, para graficar
from scipy.spatial.distance import cdist # importamos esta libreria para calcular distancias entre pares de puntos o conjuntos de puntos en un espacio vectorial

#Definimos la señal 1

w1 = 1000*np.pi #velocidad angular
f1 = w1/(2*np.pi)  #frecuencia
T1 = 1/f1  #periodo
A1 = 3  #amplitud

#Definimos la señal 2

w2 = 2000*np.pi  #velocidad angular
f2 = w2/(2*np.pi)  #frecuencia
T2 = 1/f2  #periodo
A2 = 5  #amplitud

#Definimos la señal 3

w3 = 11000*np.pi  #velocidad angular
f3 = w3/(2*np.pi)  #frecuencia
T3 = 1/f3  #periodo
A3 = 10  #amplitud

Fs = 5000  #frecuencia propuesta en el ejercicio
Ts = 1/Fs  #Periodo, se da ya que el periodo es el inverso a la frecuencia
tv = np.arange(0,2*T1, Ts) #procesa dos periodos discretizados
#inicio en 0 y finaliza dos veces T1, dando saltos cada Ts, esto sigue el periodo de la señal más lenta
y = 13*np.cos(w1*tv) + A2*np.sin(w2*tv)  #al discretizar la función original, encontramos que la señal 3 es un aliasing de la señal 1, por lo tanto al tener la misma frecuencia angular podemos sumar las dos señales

#Mostramos en pantalla la señal
plt.plot(tv,y)  #señal graficada
plt.show()  #mostrar señal

#Cero y pendiente
ycs = my_ceropen(y, ymin=-5,ymax=5) #escalamos'y' con los valores que va tomando la pendiente en cada recta, establecemos los limites en el eje 'y', la amplitud más alta es de valor 10, por lo tanto la señal va de -5 a 5 y asi tenemos 10 unidades

#Número de bits y vector de cuantización
nbits = 3
rmin = -5
rmax = 5
ve = np.linspace(rmin,rmax,2**nbits) #si se requiere incluir rmin y rmax --> (rmax-rmin)/(2^#bits-1)

#distancia y cero pendiente a vector cuantización
d = cdist(ycs.reshape(-1,1),ve.reshape(-1,1)) #se utiliza reshape(-1,1) para generar vectores columna como lo pide cdist
d.shape # d es una matriz de 500 x 8, ej: la fila 1 incluye la distancia euclidea entre y[0] y los 8 elementos en (ve)
```

Recorremos el vector cero y la señal discretizada para cuantizar

```python
for i in range(ycs.shape[0]):  #para cada i que este en el rango ()
  if i==0:  #muestra 1
    yq = my_cuantizador(ycs[i],ve)
  else:
    yq = np.append(yq,my_cuantizador(ycs[i],ve)) # si i>0 aplicar datos cuantizados
```

C. Implemente una simulación para encontar la salida del sistema lineal e invariante al tiempo $\mathcal{H}\{\cdot\}$, con respecto al escalón $h_\epsilon[n]=\{2,4,1,5,0,10\}$, $n\in\mathbb{z}$, ante la entrada análoga en corriente $x(t)=20(\cos(t/3)+\cos(t/4)) [A]$. A: Amperios. Incluya los acondicionaminetos necesarios de discretización y cuantizaión, asumiendo un microprocesador de 4 bits con entrada análoga de 4mA a 20 mA.

##Respuesta punto (C)

Revisando la señal $x(t)$, se tiene que:

$\frac{\omega_1}{\omega_2} =\frac{1/3}{1/4} = \frac{4}{3} \in \mathbb{Q}$; por ende $x(t)$ es cuasiperiódica, con período:

$T_1 = \frac{2\pi}{\omega_1} = \frac{2\pi}{1/3} = 6\pi$

$T_2 = \frac{2\pi}{\omega_2} = \frac{2\pi}{1/4} = 8\pi$

$T = T_1 l = T_2 k$

$T = 6\pi l = 8\pi l$

con $k,l\in \mathbb{Z}.$ El $MCM(6,8)=24,$ así:

$\boxed{T = 24\pi [s]}$

Además, se requiere aplicar cero y pendiente:

$\hat{x}(t)=m x(t)+ b$, con:

$m = \frac{\hat{x}_{max}(t)-\hat{x}_{min}(t)}{x_{max}(t)-x_{min}(t)}$

$b = \hat{x}_{min}(t)-m x_{min}(t)$

```python
#función cero y pendiente
def my_ceropend(x,xhmax=0.02,xhmin=0.004):
  #cálculo de pendiente y cero
  m = (xhmax-xhmin)/(x.max()-x.min())
  b = xhmin - m*x.min()
  xh = m*x+b
  return xh
```

Se implementan los códigos de discretización y cuantización:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

T1 = 6*np.pi #periodo señal 1
T2 = 8*np.pi #periodo señal 2
F1 = 1/T1
F2 = 1/T2
T = 24*np.pi #perido cuasiperiódica
Fs = 10*max(F1,F2) #frecuencia muestreo
t = np.arange(0,3*T,1/Fs) #vector de tiempo, se muestran 3 periodos
x = 20*(np.cos(t/3)+np.cos(t/4))
xhmax = 0.020
xhmin = 0.004
xh = my_ceropend(x,xhmax, xhmin).reshape(-1,1)#aplicar cero y pendiente de 0.004 a 0.020
#asegurar vector columna

#cuantizar
nbits = 4 #número bits
Ne = 2**nbits #cantidad de estados
res = (xhmax-xhmin)/Ne #resolución cuantización
ve = np.arange(xhmin,xhmax,res).reshape(-1,1)#vector de estados
#asegurar vector columna

xq = np.zeros(len(xh)) #señal cuantizada
for i in range(len(xh)): #recorrer señal cero y pendiente
    xq[i] = ve[np.argmin(cdist(xh[i].reshape(-1,1),ve))] #se cálcula la distancia desde cada punto al vector de estados

#gráficas de señales
plt.plot(t,x,'r',label='$x(t)$')
plt.xlabel('t [s]')
plt.ylabel('Amplitud')
plt.show()
plt.plot(t,xh,'b',label='$\hat{x}(t)$')
plt.stem(t,xq,'y',label='$x_{q}[n]$')
plt.xlabel('n [muestra]')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```

Teniendo la señal discretizada y cuantizada, se procede a encontrar la salida del SLIT aplicando la convolución entre la entrada digital y la respuesta impulso. Dado que tenemos la respuesta escalón, calculamos la respuesta impulso como:

$h[n] = {\rm{diff}}[h_\epsilon[n]] = h_\epsilon[n]-h_\epsilon[n-1]$

```python
he = np.array([2,4,1,5,0,10]) #respuesta al escalón del slit
h = np.diff(he)#respuesta al impulso del slit
print(h)
```

La salida del sistema se puede obtener entonces mediante la operación de convolución:

$y[n]=x[n]*h[n]=\sum_{k=-\infty}^{\infty}x[k]h[n-k]$

```python
y = np.convolve(xq,h,mode='same')#respuesta del slit -> se asegura tamaños iguales con mode=same
#graficar señal
plt.stem(t,y,'b',label='$y[n]$')
plt.legend()
plt.xlabel('n [muestra]')
plt.ylabel('Amplitud')
plt.show()
```
